If their is going over a certian speed they have the chance to damage their car severely. One of the best ways to find a neighborhood to speed in is finding a place that has limited speed humps.

```python
import pandas as pd
import geopandas
%matplotlib inline
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
neighborhoods.plot()
```

This is showing a map of the city of Pittsburgh with all of their neighborhoods.

```python
humps = pd.read_csv("speedHumps.csv")
# filter to important info
num_humps = humps.groupby("neighborhood").sum()['number_humps']
num_humps.sort_values(ascending=True)
```

This is grouping all of the neighborhoods from least speed humps to most speed humps using pandas. 

```python
hump_map = neighborhoods.merge(num_humps, how='left', left_on='hood', right_on='neighborhood')

hump_map.plot(column='number_humps',
             cmap='BuPu',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Number of Humps"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

This is showing a map of Pittsburgh showing which neighborhood has the most speed humps through coloring on the map. This shows that the more purple the map gets, the more speed humps they have. 

```python
num_humps.sort_values(ascending=False).plot.barh()
```
